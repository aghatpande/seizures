```python
#script to connect to the CHB-MIT Scalp EEG Database and download all the data from the folder of a given subject
# the data is saved in a folder named after the subject in '~/mne_data/MNE-chbmit-data/'
```

```python
import os
```

```python
# create a folder for the subject
subject = 'chb01'
path = os.path.join(os.path.expanduser('~'), 'mne_data', 'MNE-chbmit-data', subject)
os.makedirs(path, exist_ok=True)
```

```python
# specify the URL for the CHB-MIT dataset
base_url = 'https://physionet.org/files/chbmit/1.0.0/'

# specify the URL for the data files
data_url = f'{base_url}/{subject}/'

# use wget in the shell to download all files from the data URL
os.system(f'wget -P {path} -r -np -nH --cut-dirs=5 --reject "index.html*" {data_url}')
```
